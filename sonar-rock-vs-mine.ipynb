Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collectin and Data Processing

```python
#loading the dataset to as pandas DataFrame
sonar_data = pd.read_csv("/kaggle/input/connectionist-bench-sonar-mines-vs-rocks/sonar.all-data.csv",header=None)
```

```python
sonar_data=sonar_data.drop(index=0, axis=0)
```

```python
sonar_data[60].value_counts()
```

M -> Mine

R -> Rock

```python
#separating data and labels
X=sonar_data.drop(columns=60, axis=1)
Y=sonar_data[60]
```

```python
print(X)
print(Y)
```

**Training and Test data**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, stratify=Y, random_state=1)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
print(X_train)
print(Y_train)
```

**Model Training --> Logistic Regression**

```python
model = LogisticRegression()
```

```python
#training the Logisitic Regression model with training data
model.fit(X_train, Y_train)
```

**Model Evaluation**

```python
#accuracy on the training Data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print("Accuracy on training data : " , training_data_accuracy)
```

```python
#accuracy on the training Data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print("Accuracy on test data : " , test_data_accuracy)
```
